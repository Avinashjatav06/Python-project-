# 15)	Give the complete timeline of Twitter in the stock market. (Line Graph)

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Open'], label='Twitter Stock Price', color='blue')
plt.xlabel('Year')
plt.ylabel('Open')
plt.title('Twitter Stock Price Timeline')
plt.legend(loc='upper left')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 16)	Show the various types of Series from the given data. ( 3D Scatter Plot Graph.)

```python
%matplotlib notebook
plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')

fg=ax.scatter3D(df['Open'],df['High'],df['Close'],c=df['Low'],
                alpha=0.4)                
ax.set_xlabel("Open")
ax.set_ylabel("High")
ax.set_zlabel("Close")
plt.colorbar(fg)
                
```

# 18)	Show the quartile, inter-quartile Lower, Median, and Upper Quartile on the given dataset  you can use any three columns for that (Bod-Plot)

```python
np.percentile(df['Open'], [25,50,75])
np.percentile(df['Close'], [25,50,75])
np.percentile(df['High'], [25,50,75])
np.percentile(df['Low'], [25,50,75])
```

```python
print(df['Open'].median())
print(df['Close'].median())
print(df['High'].median())
print(df['Low'].median())
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(column=['Open', 'High', 'Close'])
plt.title('Box Plot with Quartiles')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()

# Calculate quartiles
quartiles = df.quantile([0.25, 0.5, 0.75])
print("Quartiles:")
print(quartiles)
```
